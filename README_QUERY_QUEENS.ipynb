#  Yanelis Music Awards (20102024)

Proyecto de anlisis musical que convierte datos de Spotify y Last.fm en una gala de premios ficticia: **los Yanelis Awards**.  
Analizamos ms de *X* canciones de **Reggaeton, Pop Latino, Funk, Rock Indie y Jazz** entre 2010 y 2024 para decidir ganadores como lbum atemporal, Artista Top Hits o lbum ms slido.

---

##  Objetivo del proyecto

- Construir una **base de datos relacional** en MySQL con informacin de canciones, artistas, lbumes y gneros.  
- Integrar datos procedentes de **mltiples CSV** (Spotify + Last.fm) y normalizarlos en un modelo 3FN.  
- Disear una **gala de premios musicales** basada en mtricas objetivas (popularidad, playcount, oyentes, etc.).  
- Practicar un flujo real de trabajo con **Python (pandas) + SQL + Visual Studio Code + Git/GitHub**.

---

##  Arquitectura de datos

La base de datos `musica_db` contiene 4 tablas principales:

- **`artistas`**  
  - `Artista` (PK)  
  - `Playcount_LastFM` (BIGINT)  
  - `Listeners_LastFM` (BIGINT)  
  - `Biografia_Resumen` (TEXT)

- **`albumes`**  
  - `ID_Album` (PK)  
  - `Nombre_Album`  
  - `Ao_Lanzamiento_Album`  
  - `Artista` (FK  `artistas`)

- **`generos`**  
  - `genero_id` (PK)  
  - `nombre_genero`

- **`canciones`**  
  - `ID_Spotify` (PK)  
  - `Nombre`  
  - `Ao_lanzamiento`  
  - `ID_Album` (FK  `albumes`)  
  - `Artista` (FK  `artistas`)  
  - `genero_id` (FK  `generos`)  
  - `Popularidad` (INT)

Este modelo permite responder preguntas como:

- Qu artista domina cada gnero entre 2010 y 2024?  
- Qu lbum es el ms consistente en popularidad?  
- Qu canciones son las ms escuchadas segn Last.fm?

##  Ejemplos de premios calculados

- **lbum atemporal 20102024**  lbum con popularidad sostenida a lo largo del periodo.  
- **Top Hits 20102024**  artista con mayor suma de popularidad / playcount.  
- **lbum ms slido**  lbum con mayor popularidad media de sus canciones.  
- **lbum ms variado**  lbum con mayor diversidad de gneros o colaboraciones.  
- **Gnero del ao**  gnero con mayor popularidad media en cada ao.
